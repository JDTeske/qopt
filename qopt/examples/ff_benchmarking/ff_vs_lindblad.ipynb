```python

import numpy as np
from qopt.examples.ff_benchmarking.singlet_triplet_setup import \
    create_simulators, create_optimizer
from qopt.analyser import Analyser
from qopt.data_container import DataContainer
```

```python
simulator_ff, simulator_l, simulator_syst = create_simulators()

syst_opt = create_optimizer(simulator_syst)
l_opt = create_optimizer(simulator_l)
ff_opt = create_optimizer(simulator_ff)
```

```python
np.random.seed(0)

initial_conditions = np.random.rand(10, 1)

data = DataContainer()
result = syst_opt.run_optimization(
    initial_control_amplitudes=initial_conditions)

data.append_optim_result(result)
analyser = Analyser(data=data)
analyser.plot_costs()
```

```python
syst_opt.system_simulator.solvers[0].plot_bloch_sphere()
```

```python
np.random.seed(0)

initial_conditions = np.random.rand(10, 1)
data = DataContainer()
result = l_opt.run_optimization(
    initial_control_amplitudes=initial_conditions)

data.append_optim_result(result)
analyser = Analyser(data=data)
analyser.plot_costs()
```

```python
np.random.seed(0)
initial_conditions = np.random.rand(10, 1)
data = DataContainer()
result = ff_opt.run_optimization(
    initial_control_amplitudes=initial_conditions)

data.append_optim_result(result)
analyser = Analyser(data=data)
analyser.plot_costs()
```
